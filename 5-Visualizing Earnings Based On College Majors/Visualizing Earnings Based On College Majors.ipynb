# Pandas, Histograms

```python
recent_grads['Sample_size'].plot(kind='hist')
```

```python
recent_grads['Sample_size'].hist(bins=25, range=(0,5000))
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(6, 12))
for r in range(1, 5):
    ax = fig.add_subplot(4, 1, r)
    ax = recent_grads[cols[r]].plot(kind = 'hist', rot = 25)
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(6, 12))
for r in range(4, 8):
    ax = fig.add_subplot(4, 1, r-3)
    ax = recent_grads[cols[r]].plot(kind = 'hist', rot = 25)
```

## Men vs Women for Major Categories

```python
tbl = recent_grads.pivot_table(index = 'Major_category', values = ['Men', 'Women'], aggfunc = sum)
tbl.plot.bar(x = tbl.index, y = ['Men', 'Women'], figsize = (10, 8))
```
